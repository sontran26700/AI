```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[2,4.1,6.3,10.2,10.5,12.1,14.1]

plt.plot(x,y,'ro')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# tính hàm loss 
def loss(x,y,w):
    l = 0
    for i in range(len(x)):
        l += (y[i] - (w[0]*x[i] + w[1]))**2
    return l/(2*len(x))

# hàm y = f(x)   
# w: tham số 
def f(x,w):
    return w[0]*x+w[1]  

# tính đạo hàm tại điểm w[0] và w[1]
def derivative(x,y,w):
    d0=0
    d1=0
    for i in range(len(x)):
        d0 += x[i]*(f(x[i],w)-y[i])
        d1 += f(x[i],w)-y[i]
    return d0*w[0]/len(x),d1/len(x)   
```

```python
# training 
epoch = 10
learning_rate = 0.01
w = [1,1] # y = x + 1
los_old = 0
for i in range(epoch):  
    # hiển thị đồ thị 
    plt.plot(x,y,'ro')
    plt.xlabel('X')
    plt.ylabel('Y')
    x0 = np.linspace(start=1, stop=10, num=50)
    y0 = w[0]*x0+w[1]
    plt.plot(x0,y0)
    plt.show()
    
    # cập nhật tham số 
    los = loss(x,y,w)
    print(los,' : ',los_old)
    if los>(los_old-0.0001) and i>0:
        break
    los_old = los
    # cập nhật 
    a,b = derivative(x,y,w)
    w[0] = w[0]-a*learning_rate
    w[1] = w[1]-b*learning_rate
    
    
    
```
